```python
from ultralytics import YOLO
import ultralytics.engine.results as yolo_results

import matplotlib.pyplot as plt
import numpy as np

# Load a model
# model = YOLO("models/yolo11n-seg.pt")  # load an official model
model = YOLO("models/n/20241105.pt")  # load a custom model

# Predict with the model
test = r'F:\codes\pl\plot-labeling\indexed\211\origin.png' #"test_imgs/1.png"
results: list[yolo_results.Results] = model(test)  # or "data/images", "test.jpg", "test.png", etc.

print(len(results))
# Process results list
result = results[0]  # Results object for the first image
boxes = result.boxes  # Boxes object for bounding box outputs
masks = result.masks  # Masks object for segmentation masks outputs
keypoints = result.keypoints  # Keypoints object for pose outputs
probs = result.probs  # Probs object for classification outputs
obb = result.obb  # Oriented boxes object for OBB outputs

""""""
df = result.to_df()  # Convert results to a pandas DataFrame
test1 = plt.imread(test)
msks = masks.data.numpy()
msks_normalized = masks.xyn
#print(masks.orig_shape, msks.shape, test1.shape)
import cv2
for i, msk in enumerate(msks):
    plt.imshow(test1)

    # Ensure the mask is in the correct format
    msk_uint8 = (msk * 255).astype('uint8')
    msk_uint8 = cv2.resize(msk_uint8, (test1.shape[1], test1.shape[0]), interpolation=cv2.INTER_CUBIC)
    contours, _ = cv2.findContours(msk_uint8, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    plt.imshow(msk_uint8,alpha=0.6)

    seg_dict = df["segments"].iloc[i]
    xs = seg_dict["x"]
    ys = seg_dict["y"]
    # plt.plot(xs,ys)

    # convert msks_normalized to the same format as msk
    msk2 =np.zeros_like(msk)
    for x,y in msks_normalized[i]:
        msk2[int(y * test1.shape[0]),int(x * test1.shape[1])] = 1
        print(int(y * test1.shape[0]),int(x * test1.shape[1]))
    msk2_uint8 = (msk2 * 255).astype('uint8')
    contours, _ = cv2.findContours(msk2_uint8, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    plt.imshow(msk2_uint8,alpha=0.1)
    plt.show()

result.show()  # Show results

df 
```
